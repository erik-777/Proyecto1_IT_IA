# Detección Temprana de Signos Neurocognitivos Asociados al Alzheimer con IA

## 1. Entendimiento del Negocio

### 1.1 Declaración del problema
En el departamento de Sucre, Colombia, el diagnóstico de la enfermedad de Alzheimer se realiza comúnmente en etapas avanzadas, cuando los signos clínicos ya son evidentes y el deterioro cognitivo es difícil de revertir. Esta situación limita la posibilidad de intervención temprana y seguimiento preventivo. A pesar del crecimiento poblacional de personas mayores, existen barreras tecnológicas, de acceso y diagnóstico en los territorios. El problema principal que se busca abordar es la **falta de herramientas accesibles y eficientes que permitan una detección temprana de signos neurocognitivos asociados a la enfermedad de Alzheimer**.

## 1.2 Árbol de Problemas

A continuación se presenta el árbol de problemas construido para identificar las causas principales del diagnóstico tardío de Alzheimer en adultos del departamento de Sucre:

![Árbol de Problemas](arbol_de_problemas.png)

### Explicación:
El problema central identificado es el **diagnóstico tardío de signos neurocognitivos asociados al Alzheimer**, lo cual genera consecuencias como:

- Intervención médica tardía o reactiva
- Mayor carga emocional y económica en las familias
- Mayor presión sobre el sistema de salud público

Entre las causas más relevantes se encuentran:

- **Falta de herramientas tecnológicas accesibles**
- **Escasa recolección y digitalización de datos clínicos**
- **Limitada formación tecnológica del personal médico**

Este árbol sirve como base para plantear una solución tecnológica integral basada en inteligencia artificial.

### 1.3 Objetivo SMART
**Objetivo general:**

Diseñar e implementar un modelo de machine learning que permita detectar tempranamente signos neurocognitivos asociados al Alzheimer en adultos del departamento de Sucre, utilizando variables biológicas y psicológicas, y facilitando el acceso a través de una interfaz digital con reconocimiento de voz.

**Objetivos específicos:**

1. Identificar y seleccionar las variables más relevantes que contribuyen al diagnóstico temprano de Alzheimer.
2. Entrenar y evaluar modelos de aprendizaje automático que permitan clasificar correctamente los signos neurocognitivos.
3. Desarrollar un prototipo funcional de una aplicación web con entrada por voz que permita recolectar información de los usuarios y dar una predicción.
4. Validar el desempeño del sistema en condiciones reales, siguiendo protocolos científicos y estándares de calidad.

### 1.4 Revisión de la literatura

Diversos estudios han resaltado la importancia de detectar tempranamente signos de deterioro cognitivo leve como una medida preventiva frente a enfermedades neurodegenerativas como el Alzheimer. Según Nagarajan et al. (2025), los modelos de aprendizaje profundo han demostrado ser eficaces en la identificación de patrones sutiles en datos clínicos y conductuales. Long (2023), en su informe global sobre Alzheimer, insiste en la necesidad de implementar herramientas tecnológicas accesibles y adaptables para el diagnóstico temprano, especialmente en países en desarrollo.

En el contexto colombiano, Lastre Meza (2020) y MinSalud (2024) destacan la falta de herramientas sistematizadas, la escasa digitalización de historias clínicas y la limitada formación técnica como barreras para una atención eficiente en salud mental. Por ello, el uso de inteligencia artificial y procesamiento de lenguaje natural representa una solución innovadora para mejorar el acceso y la precisión en la detección temprana del Alzheimer.

## 2. Entendimiento de los Datos

### 2.1 Visión General de los Datos Disponibles

El conjunto de datos utilizado para este proyecto proviene de una base de datos simulada con características clínicas, biológicas y conductuales relacionadas con personas adultas mayores. El archivo se encuentra en formato `.csv` y contiene un total de **2149 registros** y **35 variables**.

La base fue diseñada con el propósito de representar de manera realista las condiciones de salud de adultos potencialmente afectados por trastornos neurocognitivos como el Alzheimer, permitiendo evaluar su relación con diversos factores predictivos.

Los datos incluyen información sobre:

- Características demográficas (edad, género, etnia, nivel educativo)

El objetivo del análisis exploratorio y predictivo es determinar qué variables se relacionan más con el diagnóstico de Alzheimer (`Diagnosis`), representado como una variable binaria (0: no diagnosticado, 1: diagnosticado).

#### Selección de variables clave para el modelo

```python
df_clean = df.drop(['PatientID', 'DoctorInCharge'], axis=1)
```

```python
# Definir las variables predictoras y la variable objetivo
selected_features = [
    'FunctionalAssessment', 'ADL', 'MemoryComplaints', 'MMSE', 'BehavioralProblems',
    'SleepQuality', 'EducationLevel', 'CholesterolHDL', 'Hypertension',
    'FamilyHistoryAlzheimers'
]

X = df_clean[selected_features]  # Datos de entrada
y = df_clean['Diagnosis']        # Variable objetivo
```

#### Verificar balance, escalar y dividir datos

```python
print("Distribución de clases en Diagnosis:")
print(y.value_counts(normalize=True))
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
print(f"\nTamaño del conjunto de entrenamiento: {X_train.shape[0]}")
print(f"Tamaño del conjunto de prueba: {X_test.shape[0]}")
```

#### Entrenamiento del modelo

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Reporte de clasificación:\n")
print(classification_report(y_test, y_pred))
print("Matriz de confusión:")
print(confusion_matrix(y_test, y_pred))
```
